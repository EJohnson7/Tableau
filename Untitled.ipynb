```python
extension = 'zip'
filenames = []
filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
filenames
```

```python
csvs = []

for i in filenames:
    j = i.replace('.zip','')
    csvs.append(j)

print(csvs)
```

```python
combined = pd.DataFrame()
for filename in filenames:
    zf = zipfile.ZipFile(f"C:\\Users\ejohn\Desktop\Data-Homework\Tableau\{filename}") 
    csv = filename.replace('.zip','')
    df = pd.read_csv(zf.open(csv))
    combined = pd.concat(df, ignore_index = True)
```
